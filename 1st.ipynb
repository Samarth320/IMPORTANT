```python
print("replacing the missing values in gender and ssc_p:\n")
df["gender"].fillna(df["gender"].mode()[0] , inplace = True)
df["ssc_p"].fillna(df["ssc_p"].mean() , inplace = True)
```

```python
print("changing data types of columns\n")
df.dtypes
```

```python
df["sl_no"] = df["sl_no"].astype("int8")
df["sl_no"].dtypes
df
```

```python
print("converting categorical method into quantitative:\n")
df["gender"].replace({"M":1 , "F":0 } , inplace = True)
```

```python
df
```

```python
print("normalization \n")
df["salary"] = (df["salary"] - df["salary"].min() ) / ( df["salary"].max() - df["salary"].min() )
df.head()
```
